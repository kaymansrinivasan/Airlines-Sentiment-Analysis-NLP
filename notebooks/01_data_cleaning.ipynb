## Data Cleaning

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\User\airline-sentiment\data\Airline Review.csv")
print(df.head())
```

```python
df.info()
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df.drop(columns=['Aircraft'], inplace=True)
```

```python
df["Overall_Rating"] = pd.to_numeric(df["Overall_Rating"], errors='coerce')
df['Review Date'] = (
    df['Review Date']
    .str.strip()
    .str.replace(r'(\d{1,2})(st|nd|rd|th)', r'\1', regex=True)
)

df['Review Date'] = pd.to_datetime(
    df['Review Date'],
    format='%d %B %Y',
    errors='coerce'
)

df['Date Flown'] = pd.to_datetime(
    df['Date Flown'].astype(str).str.strip(),
    format='%B %Y',
    errors='coerce'
)
```

```python
print(df['Review Date'].dropna().head(10).tolist())
```

```python
print(df['Date Flown'].dropna().head(10).tolist())
```

```python
print(df.isna().sum())
```

```python
df = df.dropna(subset=['Overall_Rating'])
print(df.isna().sum())
```

```python
aspect_cols = ['Seat Comfort', 'Cabin Staff Service', 'Food & Beverages',
               'Ground Service', 'Value For Money']
df[aspect_cols] = df[aspect_cols].fillna(df[aspect_cols].median())
```

```python
df.isna().sum().sort_values(ascending=False)
```

```python
df['Type Of Traveller'] = df['Type Of Traveller'].fillna('Unknown')
df['Seat Type'] = df['Seat Type'].fillna('Unknown')
df['Route'] = df['Route'].fillna('Unknown')
```

```python
df.isna().sum().sort_values(ascending=False)
```

```python
df["Wifi & Connectivity"].value_counts().sort_values()
```

```python
df.drop(columns=['Wifi & Connectivity'], inplace=True)
```

```python
df["Inflight Entertainment"].value_counts().sort_values()
```

```python
df.drop(columns=['Inflight Entertainment'], inplace=True)
```

```python
df.isna().sum().sort_values(ascending=False)#Date Flown is left in the special case
```

```python
print(df.shape)
df.info()
```

```python
print(df.head())
```

```python
def rating_to_sentiment(r):
    if r <= 4: return 'negative'
    elif r <= 6: return 'neutral'
    else: return 'positive'

df['sentiment'] = df['Overall_Rating'].apply(rating_to_sentiment)
```

```python
df['label'] = df['sentiment'].map({
    'negative': 0,
    'neutral': 1,
    'positive': 2
})
```

```python
df['full_text'] = df['Review_Title'] + " " + df['Review']
```

```python
df['recommended_binary'] = df['Recommended'].str.lower().map({
    'yes': 1,
    'no': 0
})
```

```python
df['Review_Title'] = df['Review_Title'].str.replace('"', '').str.strip()
df['Review'] = df['Review'].str.replace('"', '').str.strip()
```
